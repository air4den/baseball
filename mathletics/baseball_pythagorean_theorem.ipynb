# Chapter 1 - Mathletics - Baseball's Pythagorean Theorem
Bill James win% estimation formula:
$$
\frac{\text{runs socred}^2}{\text{runs socred}^2 + \text{runs allowed}^2}
$$

- More runs scored increases predicted win%
- Fewer runs allowerd increases win%

$$
R = \frac{\text{runs scored}}{\text{runs allowed}}
$$



```python
from pybaseball import team_batting
from pybaseball import standings
import pandas as pd

start_year = 2005
```

```python
logs = standings(start_year)
batting_data = team_batting(start_year)
```

```python
df = pd.concat(logs)
```

```python
batting_data[['Season', 'Team', 'R']]
```
